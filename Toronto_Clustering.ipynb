Import BeautifulSoup library and use it for web scraping table from Wikipedia

```python
from bs4 import BeautifulSoup

print("Done.")
```

```python
source = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

Convert HTML table to Pandas dataframe

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

Data processing and cleaning - removing rows with 'not assigned', combining rows with same postal code and separating with commas, and replacing non-assigned neighborhoods with borough names

```python
# Dropping the rows where Borough is 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

# Combining the neighborhoods with same Postalcode
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# Replacing the name of the neighborhoods which are 'Not assigned' with names of Borough
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])

df2
```

Use shape method to print the number of rows

```python
df2.shape
```

# Part 2: Import and append latitude and longitude for each borough

First, import the csv file containing the lats and longs for the neighborhoods, saving as a pandas dataframe

```python
#read in the file and save as a pandas dataframe named lat_lon
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

Merge the two tables, matching on the postal codes

```python
df3 = pd.merge(df2,lat_lon,on='Postal Code')
df3
```

# Part 3: Clustering and Plotting the neighborhoods

First, use geopy library to get the latitude and longitude values of Toronto.

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="tor_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto is {}, {}.'.format(latitude, longitude))
```

Then make a new dataframe that only has the rows with boroughs having 'Toronto' in the name. 

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4.head()
```

```python
df4.shape
```

Visualize the neighborhoods, plotting them on a map of Toronto.

```python
map_toronto = folium.Map(location=[latitude,longitude],zoom_start=10)

for lat,lng,borough,neighborhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)

map_toronto
```

```python
# find the number of and names of boroughs
df4.Borough.unique()
```

Now, use **KMeans** to cluster the neighborhoods. I will cluster/group by location (latitude and longitude) [since it's basically the only info we have in the database right now]. Since there are 4 boroughs, I will use k=4 to see if the algorithm groups them by borough (are boroughs grouped "properly" by location?)

```python
#create dataframe of just latitudes and longitudes, fit to the kMeans model 
#then add the cluster label back into the original dataframe 

k=4
toronto_clustering = df4.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df4
```

Map the clusters

```python
# create map
map_clusters = folium.Map(location=[latitude,longitude],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['Cluster Labels']):
    label = folium.Popup('Cluster '+str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

The neighborhoods look to be clearly segmented by North, Central, East and West.
